```
flights %>% group_by(year, month, day) %>% 
summarise(total_delay = sum(dep_delay, na.rm = TRUE)) %>% head()
```

```
flights %>% group_by(dest) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x=dist, y=count)) + geom_point() + geom_smooth(se = F)
```

```
flights %>% group_by(dest) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x=dist, y=count)) + geom_point(aes(color=dest))
```

```
flights %>% group_by(dest, origin) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x=dist, y=count)) + geom_point(aes(color=origin))
```

```
flights %>% group_by(dest, origin) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x = dist, y = delay)) + geom_point(aes(size = count, color = dest)) + 
facet_wrap(~ origin, nrow=2)
```

```
flights %>% group_by(dest, origin) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x = dist, y = delay)) + geom_point(aes(size = count)) + 
facet_wrap(~ dest, nrow=10)
```

```
flights %>% group_by(dest, origin) %>% 
  summarise(count = n(), dist = mean(distance, na.rm = TRUE), delay = mean(dep_delay, na.rm = TRUE)) %>%
  filter(count>20, dest != 'ABQ') %>%
ggplot(mapping = aes(x = dist, y = delay)) + geom_point(aes(size = count)) + theme_bw() + 
theme(axis.line = element_line(size = 2, color = 'red'), panel.grid.minor = element_blank(),
panel.border = element_blank(), text = element_text(size=20))

```
